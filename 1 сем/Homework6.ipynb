```python
import telebot
import urllib.request as urllib2

bot = telebot.TeleBot('прдлоьдлт', parse_mode=None)

@bot.message_handler(commands=['start', 'help'])
def send_welcome(message):
	welcome_text = 'Привет! Отправь мне ссылку на сайт в формате URL и Я проверю его доступность'
	bot.send_message(message.chat.id, welcome_text)

@bot.message_handler(content_types=['text'])
def check_url_from_message(message):
	url = message.text
	try:
		urllib2.urlopen(url)
		bot.send_message(message.chat.id, 'URL доступен. \U00002705');
	except ValueError:
		bot.send_message(message.chat.id, 'URL сформирован не верно. \U00002B55');
	except urllib2.URLError:
		bot.send_message(message.chat.id, 'URL не доступен. \U0000274C');

bot.polling()
```

```python
import telebot
import re
from itertools import islice

bot = telebot.TeleBot('онриомго', parse_mode=None)

unions = [
	'а', 'а вдобавок', 'а именно', 'а также', 'а то', 'благодаря тому что', 'благо', 'буде', 'будто', 'вдобавок', 'в результате чего', 'в результате того что', 'в связи с тем что', 'в силу того что',
	'в случае если', 'в то время как', 'в том случае если', 'в силу чего', 'ввиду того что', 'вопреки тому что', 'вроде того как', 'вследствие чего', 'вследствие того что', 'да вдобавок', 'да еще', 'да',
	'да', 'да и', 'да и то', 'дабы', 'даже', 'даром что', 'для того чтобы', 'же', 'едва', 'ежели', 'если', 'если бы', 'затем чтобы', 'затем что', 'зато', 'зачем', 'и', 'и все же', 'и значит', 'а именно',
	'и поэтому', 'и притом', 'и все-таки', 'и следовательно', 'и то', 'и тогда', 'и еще', 'ибо', 'и вдобавок', 'из-за того что', 'или', 'кабы', 'как', 'Как скоро', 'как будто', 'как если бы', 'как словно',
	'как только', 'как-то', 'когда', 'коли', 'к тому же', 'кроме того', 'либо', 'лишь', 'лишь бы', 'лишь только', 'между тем как', 'нежели', 'невзирая на то что', 'независимо от того что',
	'несмотря на то что', 'но', 'однако', 'особенно', 'оттого', 'оттого что', 'отчего', 'перед тем как', 'по мере того как', 'по причине того что', 'подобно тому как', 'пока', 'покамест', 'покуда',
	'пока не', 'после того как', 'поскольку', 'потому', 'потому что', 'почему', 'прежде чем', 'при всем том что', 'при условии что', 'притом', 'причем', 'пускай', 'пусть', 'ради того чтобы', 'раз',
	'раньше чем', 'с тем чтобы', 'с тех пор как', 'словно', 'так как', 'так что', 'также', 'тем более что', 'тогда как', 'то есть', 'тоже', 'только', 'только бы', 'только что', 'только лишь',
	'только чуть', 'точно', 'хотя', 'чем', 'что', 'чтоб', 'чтобы'
]

pretexts = [
	'без', 'безо', 'близ', 'в', 'во', 'вместо', 'вне', 'для', 'до', 'за', 'из', 'изо', 'из-за', 'из-под', 'к', 'ко', 'кроме', 'между', 'меж', 'на', 'над', 'надо', 'о', 'об', 'обо',
	'от', 'ото', 'перед', 'передо', 'пред', 'предо', 'пo', 'под', 'подо', 'при', 'про', 'ради', 'с', 'со', 'сквозь', 'среди', 'у', 'через', 'чрез'
]

exclusion_set = set(unions + pretexts)

@bot.message_handler(commands=['start', 'help'])
def send_welcome(message):
	welcome_text = 'Привет! Отправь мне текст, а я проведу его анализ.'
	bot.send_message(message.chat.id, welcome_text)

@bot.message_handler(content_types=['text'])
def send_stat(message):
	word_list = list_of_words(message.text)
	count_map = word_count_map(word_list)
	unique_count = collect_unique_count(count_map)
	most_pop_count = collect_most_popular(count_map, 10)
	words_cnt = get_word_count(word_list)
	bot.send_message(message.chat.id, f'Общее колличество слов: {words_cnt} ')
	bot.send_message(message.chat.id, f'Колличество уникальных слов: {unique_count}')
	bot.send_message(message.chat.id, 'Топ 10 самых частых:')
	str = ''
	for kye, value in most_pop_count:
		str = str + f'\'{kye}\' - {value};\n'
	bot.send_message(message.chat.id, str)

def list_of_words(text):
	return re.findall(r'\w+', text)

def word_count_map(word_list):
	count_map = {}
	for word in word_list:
		if word.lower() not in exclusion_set:
			if word in count_map:
				count_map[word] = count_map[word] + 1
			else:
				count_map[word] = 1
	return count_map

def collect_unique_count(count_map):
	i = 0
	for value in count_map.values():
		if value == 1:
			i = i + 1
	return i


def collect_most_popular(count_map, n):
	sort = sorted(count_map.items(), key=lambda x: x[1], reverse=True)
	return list(islice(sort, n))

def get_word_count(word_list):
	return len(word_list)


bot.polling()
```

```python
import telebot
from telebot.types import InlineKeyboardMarkup, InlineKeyboardButton

bot = telebot.TeleBot('aksjnhfcojnhw', parse_mode=None)

@bot.message_handler(commands=['start', 'help'])
def send_welcome(message):
	welcome_text = 'Привет! Введи команду /calc и начнем расчеты.'
	bot.send_message(message.chat.id, welcome_text)

first = 0.0
action = ''
second = 0.0

@bot.message_handler(commands=['calc'])
def answer(message):
	bot.send_message(message.chat.id, 'Калькулятор', reply_markup=draw_calc())

def is_float(string):
	try:
		float(string)
		return True
	except ValueError:
		return False


@bot.callback_query_handler(func=lambda call: is_float(call.data))
def add_next_num(call):
	global first
	global second
	if action == '':
		if first == 0.0:
			first = float(call.data)
		else:
			first = first * 10 + float(call.data)
	else:
		if second == 0.0:
			second = float(call.data)
		else:
			second = second * 10 + float(call.data)
	try:
		bot.edit_message_reply_markup(call.message.chat.id, call.message.message_id, reply_markup=draw_calc())
	except Exception:
		return


@bot.callback_query_handler(func=lambda call: call.data in {'+','-','x','/', '%'})
def set_action(call):
	global action
	action = call.data
	return

@bot.callback_query_handler(func=lambda call: call.data == '+/-')
def call_reverse(call):
	global first
	first = -first
	try:
		bot.edit_message_reply_markup(call.message.chat.id, call.message.message_id, reply_markup=draw_calc())
	except Exception:
		return


@bot.callback_query_handler(func=lambda call: call.data == '=')
def calculate(call):
	global action
	global first
	global second
	if action == '+':
		first += second
	elif action == '-':
		first -= second
	elif action == 'x':
		first *= second
	elif action == '/':
		first /= second
	elif action == '%':
		first = (first*second)/100
	second = 0.0
	action = ''
	try:
		bot.edit_message_reply_markup(call.message.chat.id, call.message.message_id, reply_markup=draw_calc())
	except Exception:
		return

@bot.callback_query_handler(func=lambda call: call.data == 'ac')
def call_ac(call):
	global first
	global action
	global second
	first = 0.0
	second = 0.0
	action = ''
	try:
		bot.edit_message_reply_markup(call.message.chat.id, call.message.message_id, reply_markup=draw_calc())
	except Exception:
		return


def draw_calc():
	markup = InlineKeyboardMarkup(row_width=4)

	result = InlineKeyboardButton(str(first), callback_data=str(first))
	if action != '':
		result = InlineKeyboardButton(str(second), callback_data=str(second))
	plus = InlineKeyboardButton('+', callback_data='+')
	minus = InlineKeyboardButton('-', callback_data='-')
	mult = InlineKeyboardButton('x', callback_data='x')
	div = InlineKeyboardButton('/', callback_data='/')

	ac = InlineKeyboardButton('ac', callback_data='ac')
	pm = InlineKeyboardButton('+/-', callback_data='+/-')
	perc = InlineKeyboardButton('%', callback_data='%')

	eq = InlineKeyboardButton('=', callback_data='=')

	one = InlineKeyboardButton('1', callback_data='1')
	two = InlineKeyboardButton('2', callback_data='2')
	three = InlineKeyboardButton('3', callback_data='3')
	four = InlineKeyboardButton('4', callback_data='4')
	five = InlineKeyboardButton('5', callback_data='5')
	six = InlineKeyboardButton('6', callback_data='6')
	seven = InlineKeyboardButton('7', callback_data='7')
	eighth = InlineKeyboardButton('8', callback_data='8')
	nine = InlineKeyboardButton('9', callback_data='9')
	o = InlineKeyboardButton('0', callback_data='0')

	markup.add(result)
	markup.add(ac, pm, perc, div)
	markup.add(seven, eighth, nine, mult)
	markup.add(four, five, six, minus)
	markup.add(one, two, three, plus)
	markup.add(o, eq)

	return markup


bot.polling()
```
