```python
#1 ex
def changeDate(sec):
    days =sec//86400
    sec=sec-days*86400
    hours = sec//3600
    sec =sec-hours*3600
    minuts = sec//60
    sec=sec-minuts*60
    print(f'{days}:{hours}:{minuts}:{sec}')

#2 ex
def calculator():
    try:
        num1 = int(input('Enter the first num:'))
        num2 = int(input('Enter the second num:'))
        sign = input('Enter the character: ')
        if sign == '+':
            res = num1+num2
            print(f'{num1} {sign} {num2} = {res}')
        elif sign == '-':
            res = num1-num2
            print(f'{num1} {sign} {num2} = {res}')
        elif sign == '*':
            res = num1*num2
            print(f'{num1} {sign} {num2} = {res}')
        elif sign == '/' and num2!=0:
            res = num1/num2
            print(f'{num1} {sign} {num2} = {res}')
        elif sign == '/' and num2==0:
            print("You can't divide by zero")
        elif sign!='*' and sign!='/' and sign!='+' and sign!='-':
            print("Unknown character")
    except ValueError:
        print("It's not a num! Start over...")
        calculator()

#3 ex        
def merge():
    try:
        str1 = input("Enter the text: ")
        if not str1:
            raise ValueError('Empty line, enter again please...')
        str2 = input("Enter the text again: ")
        if not str2:
            raise ValueError('Empty line, enter again please...')

        print(str1 + '$' + str2)
        print(str1, str2, sep='$')
    except ValueError as e:
        print(e)
        merge()

#4 ex        
def congratulate():
    try:
        s = 1
        name = input("What is your name?")
        if name.isdigit() == True:
            raise ValueError()
        age = int(input("How old are you?"))
        if age>100:
            print("Are you kidding?")
        while (s <= age):
            print(f"{s}: Happy Birthday, {name}!")
            s += 1
    except ValueError:
        print("It's not true! Try again...")
        congratulate()

if __name__ == '__main__':
    seconds = int(input())
    changeDate(seconds)
    calculator()
    merge()
    congratulate()


```
